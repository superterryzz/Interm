```python
import pandas as pd
import numpy as np
```

```python
# 计算汉明距离 #汉明距离对于离散型数据的距离度量有很大帮助
def hamming_distance_binary(x, y):
    return bin(x ^ y).count('1')
```

```python
def hamming_distance_matrix(data):
    num_samples = len(data)
    distance_matrix = np.zeros((num_samples, num_samples))
    
    for i in range(num_samples):
        for j in range(i + 1, num_samples):
            distance = np.sum([hamming_distance_binary(x, y) for x, y in zip(data[i], data[j])])
            distance_matrix[i, j] = distance
            distance_matrix[j, i] = distance
            
    return distance_matrix
```

```python
data = pd.read_excel(r"all_samples_clean.xlsx", sheet_name="Sheet1")
data = data.drop(['BASIC_entity_name', 'BASIC_year'], axis=1)

#将非0值转换为1
data = data.applymap(lambda x: 1 if x != 0 else 0)
```

```python
data_array = np.array(data)
```

```python
# 计算汉明距离矩阵
distance_matrix = hamming_distance_matrix(data_array)
```

```python
distance_matrix
```

```python
from sklearn.cluster import KMeans
```

```python
# 使用K均值聚类对数据进行聚类
num_clusters = 3  # 设置聚类的数量
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(distance_matrix)
```

```python
# 获取每个样本所属的簇
cluster_labels = kmeans.labels_
# 打印每个样本所属的簇
for i, label in enumerate(cluster_labels):
    print(f"Sample {i+1} belongs to cluster {label+1}")
```
