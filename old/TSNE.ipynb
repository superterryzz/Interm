```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
data = pd.read_excel(r"all_samples_clean.xlsx", sheet_name="Sheet1")
data = data.drop(['BASIC_entity_name', 'BASIC_year'], axis=1)

#将非0值转换为1
data = data.applymap(lambda x: 1 if x != 0 else 0)
```

```python
data.values
```

```python
# 创建t-SNE对象并进行降维
n_components = 2 # 设置降维后的维度
tsne = TSNE(n_components=n_components, perplexity=30, n_iter=300)  
reduced_data = tsne.fit_transform(data.values)
```

```python
# 绘制降维后的数据
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c='b', marker='o')
plt.title('t-SNE')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```
