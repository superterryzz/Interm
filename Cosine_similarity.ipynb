```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_excel(r"all_samples_clean.xlsx", sheet_name="Sheet1")
data = data.drop(['BASIC_entity_name', 'BASIC_year'], axis=1)
```

```python
#将非0值转换为1
#data = data.applymap(lambda x: 1 if x != 0 else 0)  
```


*二分类和多分类没有区别*

```python
X =data.values
```

***计算特征之间的余弦相似度矩阵***


```python
# 将数据集转置，使得每一列代表一个特征
X_transposed = X.T
```

```python
# 计算特征之间的余弦相似度矩阵
similarity_matrix = cosine_similarity(X_transposed)
```

```python
similarity_matrix_df = pd.DataFrame(similarity_matrix, index=data.columns, columns=data.columns)
#similarity_matrix_df.to_excel(r"Cosine_similarity_matrix2.xlsx", sheet_name="Sheet1")
```

***筛选：根据相似度矩阵进行特征选择***

```python
# 创建一个空的列表来存储选定的特征索引
selected_feature_indices = []
```

```python
# 首先选择相似度最高的特征作为代表性特征
most_similar_feature_index = np.argmax(np.sum(similarity_matrix, axis=0))
selected_feature_indices.append(most_similar_feature_index)
```

```python
selected_feature_indices
```

```python
# 逐步选择与已选特征最不相似的特征
while len(selected_feature_indices) < 10:  # 选择前10个特征
    max_dissimilarity = 0
    selected_feature_index = None
    
    for i in range(X.shape[1]):
        if i not in selected_feature_indices:
            dissimilarity = np.sum(similarity_matrix[i, selected_feature_indices])
            if dissimilarity > max_dissimilarity:
                max_dissimilarity = dissimilarity
                selected_feature_index = i
    
    selected_feature_indices.append(selected_feature_index)
```

```python
# 获取选定的特征的名称
selected_feature_names = [f"{data.columns[idx]}" for idx in selected_feature_indices]
```
